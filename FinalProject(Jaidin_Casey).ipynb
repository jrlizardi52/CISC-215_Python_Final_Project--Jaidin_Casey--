```python
import random 
```

```python
# Define a function for the guessing game
def guessing_game():
    print("Welcome to the guessing game! Give me a lower and upper bound, and I'll pick a random number between them. Then you can try to guess it!")


        # Get the lower and upper bounds from the user
    lb = int(input("Enter the lower bound: ")) 
    ub = int(input("Enter the upper bound: "))

    # Have the program pick a random number between the bounds
    num_to_guess = random.randint(lb, ub) 

    # Initialize the number of attempts
    attempts = 0
    guessed_correctly = False


    # Loop until the user guesses the number
    while not guessed_correctly:
        try:
            guess = int(input("Enter your guess: "))
            attempts += 1

            if guess < num_to_guess:
                print("Too low! Try again.")
            elif guess > num_to_guess:
                print("Too high! Try again.")
            else:
                guessed_correctly = True
                print(f"Congrats! You guessed the number {num_to_guess} in {attempts} attempt(s)!")
        except ValueError:
            print("Invalid input. Please enter an integer.")



if __name__ == "__main__":
    guessing_game()
    

# Asks user if they wish to play again.
    again = input("\nDo you want to play again? (yes/no): ").lower()
    if again != "yes":
            print("Play again soon!")
          
```
